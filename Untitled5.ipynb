```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Quarter_supplementary_table.csv")

print("Dataset preview:")
print(df.head())
```

```python
print("total no. of rows and cols: ", df.shape)
print("\nDataset Info:",df.info())
print("\nSummary Statistics:",df.describe())
```

```python
df_gdp = df[df["Series_name"] == "Expenditure GDP (quarterly)"]
plt.figure(figsize=(8,5))
plt.plot(df_gdp["Period"], df_gdp["Value"], marker="o")
plt.title("Quarterly GDP Over Time")
plt.xlabel("Period")
plt.ylabel("Value (NZDm)")
plt.show()
```

```python
latest_period = df["Period"].max()
df_latest = df[df["Period"] == latest_period]

plt.figure(figsize=(8,5))
df_latest.groupby("Group")["Value"].sum().plot(kind="bar")
plt.title(f"Values by Group (Period: {latest_period})")
plt.ylabel("Total Value")
plt.show()
```

```python
plt.figure(figsize=(6,6))
df_latest.groupby("Group")["Value"].sum().plot(kind="pie", autopct='%1.1f%%')
plt.title(f"Group Contribution (Period: {latest_period})")
plt.ylabel("")
plt.show()
```
